```python
import pandas as pd
import numpy as np
import random
import os
import datetime
import string
```

```python
# import data
data = pd.read_csv("/home/bluepi/Downloads/Update/product_info/main table/product_info.csv")
new_product = pd.read_csv("/home/bluepi/Downloads/Update/product_info/main table/new_added.csv")
```

## Creating date named directory

```python
# Get the current date
current_date = datetime.datetime.today()

# datetime.timedelta(days=1) ---> in order to get the previous date
# strftime ('%d-%m-%Y')  ---> To get the date in time desired format
date_folder = (current_date - datetime.timedelta(days=1)).strftime ('%d-%m-%Y')

# Parent Directory path 
parent_dir = "/home/bluepi/Downloads/Update/product_info"
# os.path.join --> join one or more path components
path = os.path.join(parent_dir, date_folder) 
# Creating Directory
os.mkdir(path) 
```

## Creating CSV file

```python
value = random.randint(1,10)

#data_random_subset
data_random_subset = data.sample(n=value)

# Create the file name as it would be random
file_name = str(''.join(random.choices(string.ascii_uppercase + string.digits, k = 10) ))

# File path
path_csv = path + '/' + file_name + '.csv'
```

## Updating Data
- Removing day and time
- Adding new prices
- Adding new date_timestamp
- Adding record type -- UPDATED

```python
# Drop the existing price column
data_random_subset = data_random_subset.drop(['price','day','time'],axis=1)

# Add new price column with new prices
data_random_subset['price'] = np.random.randint(1, 1500, size=data_random_subset.shape[0])

# To create date_timestamp
# here we used list comprehensions
# datetime.datetime.today() ---> get the today's date
# datetime.timedelta(minutes= ...,seconds=...) ---> used for calculating differences in dates & date manipulations
data_random_subset['Date_timestamp'] = [(datetime.datetime.today() 
                                  + datetime.timedelta(minutes= random.randint(10,100),
                                                seconds=random.randint(10,100) )).strftime("%d/%m/%Y %H:%M:%S") 
                                     for i in range(data_random_subset.shape[0])] 

# Adding record type data
data_random_subset['record_type'] = 'U'
```

### Adding new Products

```python
value_2 = random.randint(0,5)

#new products picked at random
new_data_random_subset = new_product.sample(n=value_2)

# Updating new product file
new_product[ ~new_product['p_id'].isin(new_data_random_subset['p_id'].values) ]
```

- Adding the date timestamp
- Adding record type INSERTED

```python
# Adding the date timestamp
new_data_random_subset['Date_timestamp'] = [(datetime.datetime.today() 
                                  + datetime.timedelta(minutes= random.randint(10,100),
                                                seconds=random.randint(10,100) )).strftime("%d/%m/%Y %H:%M:%S") 
                                     for i in range(new_data_random_subset.shape[0])] 

# Adding the record type
new_data_random_subset['record_type'] = 'I'

# Concating newly added and updated products
final_day_update = pd.concat([data_random_subset,new_data_random_subset],sort = False)
```

```python
# Exporting the csv file
final_day_update.to_csv(path_csv)
```

```python
# Products which have been deleted
deleted_product = []

# Getting list of all products present in main product table
data_pid = data['p_id'].tolist()

for j in range(10):
    to_delete = random.choices( data_pid, k = random.randint(0,4) )
    print(" to be deleted --> ");print(to_delete)
    print( "All deleted products " , end= '\n') 
    print( deleted_product,end='\n')
    print(" ",end='\n')
    for i in to_delete :
        if i in deleted_product :
            print("to be removed -- >");print(i)
            to_delete.remove(i)
            print("after removing already deleted p-ID",end='\n')
            print(to_delete)
            print(" ",end = '\n')
        else :
            deleted_product.append(i)
            
```
